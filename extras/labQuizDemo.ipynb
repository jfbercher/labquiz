# Utilisation de `quiz_dash`

Au lieu de la ligne de commande, vous pouvez utiliser l'application développée spécifiquement et disponible ici :

https://jfb-quizdash.streamlit.app/

- Dans le menu de navigation, entrez les paramètres attendus : 
URL, SECRET et téléverser le fichier de questions contenant les réponses (ici note "quizzes_basic_filtering_test.yaml")

- Dans les paramètres à surveiller, entrer `{'retries':2, 'exam_mode':False, 'test_mode':False}` 
qui sont les paramètres que nous avons fixés. Toute modification de ces paramètres sera signalée. 

- Mettre le nombre d'essais autorisés à 3 (retries +1). 

👉🏼 Vous pouvez ensuite consulter les anomalies dans `Integrity live`, les résultats des participants dans `Monitoring` et enfin calculer les résultats dans l'onglet `Correction et Notes` 😊. 


>Remarque : *Le calcul de la note finale n'est réalisé *que* sur les quizzes auuxquels au moins une réponse a été apprtée dans le tableau de résultats. C'est un choix qui a été fait. Si vous voulez utiliser des questions auxquelles les élèves n'auraient pas répondu, créer un utilisateur virtuel (vous) qui répondra à toutes les questions -- ou alors je mettrai une option un jour, mais bon.* 

# Utilisation de `quiz_editor` - création d'une version cryptée, d'une version examen

On peut modifier le fichier de questions à la main, mais effectivement le plus simple est d'utiliser l'application dédiée. Rendez vous sur 

https://jfb-quizeditor.streamlit.app/

et téléversez le fichier que nous avons utilisé `quizzes_basic_filtering_test.yaml` (Commencez par sauvegarder celui-ci sur votre disque : allez dans le gestionnaire de fichier sur la gauche ou menu `View/File Browser`, sélectionner le YAML, et `File/Download`)

- Dès lors, vous pouvez éditer, ajouter ou supprimer des questions. Jouez avec cela. 

- Dans la barre de navigation de `quiz_editor`, sur la gauche, descendre vers `Exportation`. Sélectionner quelques quizzes (vous pouvez aussi les cocher "Exporter" sur la droite en les parcourant), puis choisir `Extrait YAML` dans le format et cliquer sur configurer et exporter. Choisir `YAML crypté`, un nom de fichier, et entrer `Wrktz`comme mot de passe (c'est le mot de passe qui est configuré dans la page Config du google sheet (mais comme je ne vous l'ai partagé qu'en lecture, vous ne devriez pas pouvoir le changer). Vous pouvez alors télécharger la version complète mais cryptée, et la version ``questions only'' que vous pourriez utiliser en mode examen. Mettons que vous ayez sauvegardé sous les noms `export_crypt.yaml` et `export_qo_crypt.yaml` (spécifier juste `export` dans `quiz_editor`, les suffixes et extensions sont ajoutés automatiquement).  

Il ne nous reste plus qu'à tester ceci (après avoir copié les fichiers dans votre répertoire de travail, ou téléversé dans Jupyterlite par drag and drop).
