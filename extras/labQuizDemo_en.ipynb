> Note: In this mode, the correction is calculated on all the questions in the file. If you only want to answer a few questions, cancel the excess questions via a scale. Also columns with no answers at all are not displayed. 

````python
# Construction of the new scale - 1 for the questions answered, zero otherwise
scale1 = {key:1 for key in Res.columns}
scale = {key:0 for key in quiz.quiz_bank.keys() if key not in Res.columns}
scale.update(scale1)
Res = correctQuizzes(URL, SECRET, QUIZFILE, bareme=scale)
Res
````

# Using `quiz_dash`

Instead of the command line, you can use the application developed specifically and available here:

https://jfb-quizdash.streamlit.app/

- In the navigation menu, enter the expected parameters: 
URL, SECRET and upload the question file containing the answers (here our "test.yaml")

- In the parameters to monitor, enter `{'retries':2, 'exam_mode':False, 'test_mode':False}` 
which are the parameters that we have set. Any changes to these settings will be reported. 

- Set the number of authorized attempts to 3 (retries +1). 

👉🏼 You can then consult the anomalies in `Integrity live`, the results of the participants in `Monitoring` and finally calculate the results in the `Correction and Notes` tab 😊. 


>Note: *The calculation of the final mark is carried out *only* on quizzes to which at least one answer has been included in the results table. It’s a choice that was made. If you want to use questions that students haven't answered, create a virtual user (you) who will answer all the questions -- or I'll add an option one day, but oh well.*

# Using `quiz_editor` - creating an encrypted version, an exam version

You can modify the question file by hand, but the easiest way is to use the dedicated application. Go to 

https://jfb-quizeditor.streamlit.app/

and upload the file we used `quizzes_basic_filtering_test.yaml` (Start by saving this one on your disk: go to the file manager on the left or `View/File Browser` menu, select the YAML, and `File/Download`)

- From then on, you can edit, add or delete questions. Play with this. 

- In the `quiz_editor` navigation bar, on the left, go down to `Export`. Select a few quizzes (you can also check "Export" on the right while browsing them), then choose `Extract YAML` in the format and click on configure and export. Choose `Encrypted YAML`, a file name, and enter `Wrktz` as password (this is the password that is configured in the Config page of the Google sheet (but as I only shared it with you in reading mode, you should not be able to change it). You can then download the full but encrypted version, and the ``questions only'' version which you could use in exam mode. Let's say that you have saved under the names `export_crypt.yaml` and `export_qo_crypt.yaml` (just specify `export` in `quiz_editor`, suffixes and extensions are added automatically).

All we have to do now is test this (after copying the files to your working directory, or uploading them to Jupyterlite using drag and drop).
